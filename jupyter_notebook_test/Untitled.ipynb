```python
for i in range(total_num):
    f = files[i]
    try:
        reg_info = re.findall(r'data/([A-Z]+)_([A-Z]+).csv', f)
        if reg_info:
            ticker, option_type = reg_info[0]
            if option_type == 'C':
                df = pd.read_csv(f)
                if 'downloadDate' in df.columns:
                    df = df.drop(['downloadDate'], axis=1)
                else:
                    continue
            else:
                df = pd.read_csv(f)
                if 'Ticker' in df.columns:
                    continue
                else:
                    df = pd.read_csv(f, header=None)
                    df.columns = cols[:-1]
                    df['Ticker'] = ticker
                    df['OptionType'] = option_type
            df = df.drop_duplicates(['contractSymbol', 'lastTradeDate'], keep='first')
            df['inTheMoney'] = df['inTheMoney'].map({'True':'True', 'False':'False', '1.0':'True', '0.0':'False'})
        df.to_csv(f, index=False)
    except Exception as e:
        print(f"Error happended when processing data for {f}")
        print(e)
    print(f'Finished {i}/{total_num}')
```

```python
df['contractExpiryDate'] = df['contractSymbol'].apply(lambda x: re.findall('([\d]+)[CP]', x)[0])
df['contractExpiryDate'] = '20'+ df['contractExpiryDate']
```

```python
info_company = {}
info_exchange = {}
info_type = {}
for t in df['Ticker'].unique():
    url = "http://d.yimg.com/autoc.finance.yahoo.com/autoc?query={}&region=1&lang=en".format(t)
    try:
        print(f"Started to process {t}")
        result = requests.get(url).json()
        for res in result['ResultSet']['Result']:
            if res['symbol'] == t:
                info_company[t] = res['name']
                info_exchange[t] = res['exchDisp']
                info_type[t] = res['typeDisp']
                break
    except Exception as e:
        print(f"Error happened for ticker {t}")
        print(e)
        continue
    print(f'processed ticker {t}')
```

```python
df['Company'] = df['Ticker'].map(info_company)
df['Exchange'] = df['Ticker'].map(info_exchange)
df['TypeDisp'] = df['Ticker'].map(info_type)
```
